```python
fb = pd.read_csv('data/fb_2018.csv', index_col='date', parse_dates=['date'])
fb.head()
```

```python
fb.assign(
    Z_score= lambda x: x.volume.sub(x.volume.mean()).div(x.volume.std()).abs()
).query('Z_score > 3')
```

```python
fb.assign(
    volume_pct_change = fb.volume.pct_change(),
    pct_rank = lambda x: x.volume_pct_change.abs().rank(ascending=False)
).nsmallest(5,'pct_rank')
```
